```python
#IMPORT LIBRARIES
import cv2 #openCV (C++)
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
#READ IMAGE
apolloOriginal = cv2.imread(filename=os.path.relpath('Data/Apollo_11_Launch.jpg'),
                         flags=0) #greyscale = 0, color = 1

#PRINT IMAGE INFORMATION
size = apolloOriginal.shape
print(f"Image size (H, W) is: {size}")
print(f"Data type of image is: {apolloOriginal.dtype}")

#PRINT IMAGE
plt.figure()
plt.imshow(X=apolloOriginal, cmap="gray")
```

```python
#CROP IMAGE
apolloCropped = apolloOriginal[200:500, 400:800]

#INCREASE BRIGHTNESS
apolloBrighter = cv2.add(apolloCropped, 100)
# apolloDarker = cv2.subtract(apolloCropped, 100)

#ROTATE IMAGE
apolloRotated = cv2.rotate(apolloBrighter, 0) #90 degrees clockwise = 0
                                              #180 degrees clockwise = 1
                                              #270 degrees clockwise = 2

#PRINT RESULTS
plt.figure(figsize=[10, 5]) #[width, height] (in)

plt.subplot(131); plt.imshow(X=apolloCropped, cmap="gray")
plt.title("Cropped Image")

plt.subplot(132); plt.imshow(X=apolloBrighter, cmap="gray") #notice clipping due to overflow
plt.title("Brightened Image")

plt.subplot(133); plt.imshow(X=apolloRotated, cmap="gray")
plt.title("Rotated Image")
```

```python
#SAVE IMAGE
cv2.imwrite("Data/Apollo_11_Launch_MODIFIED.jpg", apolloRotated)
```
